```python
def soma(n1, n2):
  return n1 + n2

def subtrai(n1, n2):
  return n1 - n2

def multiplicar(n1, n2):
  return n1 * n2

def divide(n1, n2):
  return n1 / n2

def potenciacao(n1, n2):
  return n1 ** n2

def raiz(n1):
  return print(f"o resultado da raiz de {n1} é:", (n1 ** (1/2)))

def maior (n1, n2):
  if n1 > n2:
    return (n1)
  else:
    return (f"{n2}")
  #elif n1 == n2:
    #return print(f"{n1} e {n2} são o mesmo valor")

def fatorial(n1):
  if n1 == 0 or n1 == 1:
    return 1
  elif n1 > 1:
    return n1 * fatorial(n1 - 1)
```
